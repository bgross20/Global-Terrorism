## 3. Cleaning Data/Consistency Checks

### Mixed Type Data

```python
## Resetting Index 
df_terror = df_terror.reset_index(drop=True)
```

```python
## Checking for mixed type data 
for col in df_terror.columns.tolist():
  weird = (df_terror[[col]].applymap(type) != df_terror[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_terror[weird]) > 0:
    print (col)
```

```python
## Seeing why they are mixed 
(df_terror[['motive']].applymap(type) != df_terror[['motive']].iloc[0].apply(type)).any(axis = 1)
```

```python
## Seeing why they are true...there are Nan values 
df_terror[106153:106158]
```

```python
## Adding missing if value is NaN
df_terror['motive'].fillna('Missing', inplace=True)
df_terror['targsubtype1_txt'].fillna('Missing', inplace=True)
df_terror['natlty1_txt'].fillna('Missing', inplace=True)
df_terror['weapsubtype1_txt'].fillna('Missing', inplace=True)
df_terror['propextent_txt'].fillna('Missing', inplace=True)
df_terror['hostkidoutcome_txt'].fillna('Missing', inplace=True)
```
